```python
df= df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name",inplace=True)
```

```python
missing_counts = df.isna().sum()
cols_with_nan = missing_counts[missing_counts > 0]
print("Columns with NaN values:")
print(cols_with_nan)
```

```python
#arranging the data in ascending order
sorting_the_data=df.sort_values(by="Publishing Year", ascending=True, inplace=True)
# Display the first 10 rows of the sorted DataFrame
df.head(10)
```

```python
df.tail(10)
```

```python
plt.hist(df["Publishing Year"], bins=10, color='green', edgecolor='black')
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df["genre"].value_counts().plot(kind='bar', color='blue', edgecolor='black')
plt.xlabel("Genre")
plt.ylabel("no _of_times_genre appears")
plt.title("Genre Distribution")
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
df.groupby("Author")['Book_average_rating'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='orange', edgecolor='black')
plt.title("top 10 Authors with highest average book rating")
plt.ylabel("Average Book Rating")
plt.xlabel("Author")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
