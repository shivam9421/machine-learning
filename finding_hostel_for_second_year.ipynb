```python
df1.rename(columns={'Unnamed: 1':'scholar_id','Unnamed: 2':'Name','Unnamed: 3':'Hostel Alloted','Unnamed: 4':"Seater"},inplace=True)
df2.rename(columns={1:'scholar_id',2:'Name',3:'Hostel Alloted',4:"Seater"},inplace=True)
```

```python
df1.dropna(inplace=True)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df.drop(columns=[0],inplace=True)
```

```python
def finding_hostelandseater_alloted(taking_input_scholar_id):

  hostel_alloted=df[df['scholar_id']==taking_input_scholar_id]
  return hostel_alloted


```
